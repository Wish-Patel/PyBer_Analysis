```python
 #Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create a plot
plt.plot(x, y)
#plt.figure(figsize=(400,800))
plt.show
```

```python
# Create the plot with ax.plt()
fig, this = plt.subplots()
this.plot(x, y)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
```

```python
#create a plot with label
plt.plot(x,y, label='Boston', marker ="*", color ="red", linewidth=2)

#create labels for x and y
plt.xlabel("Date")
plt.ylabel("Fare ($)")
#plt.subplots(figsize=(20,10))
#Set the y limit 0-45
plt.ylim(0,45)

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()

#Add a grid
plt.grid()

plt.xlim(0,5)

#plt.show()

```

```python
#create a bar chart

plt.bar(x,y, color ="green", label = "Boston")

#create labels for x and y
plt.xlabel("Date")
plt.ylabel("Fare ($)")

plt.ylim(0,45)

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
#create a bar chart

plt.barh(x,y, color ="blue", label = "Boston")
plt.gca().invert_yaxis()

#create labels for x and y
plt.ylabel("Date")
plt.xlabel("Fare ($)")



#plt.ylim(0,12)

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
#bar graph with ax.plot()
fig, ax = plt.subplots()
ax.barh(x,y)#,color="blue", label='New York')
```

```python
plt.scatter(x,y, s=y)
```

```python
larger_y = []
for data in y:
    larger_y.append(data*3)
    
plt.scatter(x,y, s=larger_y)

plt.scatter(x,y, s=[i*3 for i in y])
```

```python
#plt.pie(y, labels=x)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(12, 12))

plt.pie(
    y, 
    labels=x, 
    explode=explode_values, 
    colors = colors,
    autopct='%.1f%%'
)
plt.show()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y)
stdev
```

```python
plt.errorbar(
    x, 
    y, 
    yerr=stdev,
    capsize=3
)
```

```python
plt.bar(x, y, yerr=stdev, capsize=3)
```

```python
import numpy as np
plt.barh(x, y)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()

fig, ax = plt.subplots()
ax.barh(x, y)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x, y)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
